# Decision Tree Regression

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

## Importing the dataset

```python
df=pd.read_csv("/content/Housing.csv")
print(df.head())
```

#Preprocessing

```python
from sklearn.preprocessing import LabelEncoder
label_encoders = {}
categorical_cols = ['furnishingstatus', 'mainroad', 'guestroom', 'basement', 'hotwaterheating', 'airconditioning', 'prefarea']
for col in categorical_cols:
    le = LabelEncoder()
    df[col] = le.fit_transform(df[col])
    label_encoders[col] = le
# df['furnishingstatus']
```

```python
df.head()
```

# Seperating Features and Target

```python
X=df.iloc[:,1:]
Y=df.iloc[:,0]
```

## Splitting the dataset into the Training set and Test set

```python
from sklearn.model_selection import train_test_split
X_train, X_test, Y_train, Y_test=train_test_split(X,Y,random_state=1, test_size=0.3)
```

#Scaling the data

```python
from sklearn.preprocessing import StandardScaler
scaler=StandardScaler()
```

```python
#fitting only with trained features
scaler.fit(X_train)
```

```python
#Transforming the train and test features
# X_train=scaler.transform(X_train)
# X_test=scaler.transform(X_test)
```

```python
X_train.shape, X_test.shape
```

## Training the Simple Linear Regression model on the Training set

```python
from sklearn.tree import DecisionTreeRegressor
model=DecisionTreeRegressor()
model.fit(X_train,Y_train)
```

## Predicting the Test set results

```python
Y_pred=model.predict(X_test)
```

#Finding Accuracy

```python
from sklearn.metrics import mean_squared_error,mean_absolute_error,r2_score
mse=mean_squared_error(Y_test,Y_pred)
mae=mean_absolute_error(Y_test,Y_pred)
r2 = r2_score(Y_test, Y_pred)
print("Mean-Square Error:",mse,
      "\nMean Absolute Error:",mae,
      "\nR^2 score:",r2)
```
